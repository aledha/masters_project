```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta0 = 1
def hmarked(lmdba):
    return 1 + beta0 * (lmdba + np.minimum(lmdba, 0.87) - 1.87)

def h(lmbda):
    return np.maximum(0, hmarked(np.minimum(lmbda, 1.2)))
```

```python
lmbdas = np.linspace(0, 2, 100)
hs = h(lmbdas)
plt.plot(lmbdas, hs)
plt.xlabel(r"$\lambda$")
plt.ylabel(r"$h(\lambda)$")
plt.grid()
```
